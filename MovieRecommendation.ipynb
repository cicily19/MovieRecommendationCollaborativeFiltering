# Recommendation Systems
This are applications that are used in various platforms to suggest the likelihood of a customer preference e.g, Ecommerce we can suggest the likelihood of a customer purchasing a certain product. In news articles we can also recommend the likelihood a person reading a particular news etc. We have two major categories of recommendation systems:

## 1. Collaborative Filtering Recommendation System

Source : chatGPT

A technique used in recommendation systems to predict user's interest or preferences by collecting preferences from many users. The recommendation is done baed on oter users reviews rather than your preference.
e.g, youtube might recommend a trending video, X might recommend some trending hashtags, Netflix might recommend some TRENDING hashtag, Netflix can also recommend a movie based on its popularity.

```python
# MOVIES RECOMMENDATION USING COLLABORATIVE FILTERING
# DATA USED: IMDB MOVIES DATASET
# Users Datasets
import pandas as pd
users = pd.read_csv('https://msi.martial.co.ke/data/file.csv')
users
```

```python
# Movie IDS Dataset
movies = pd.read_csv('https://msi.martial.co.ke/data/Movie_Id_Titles.csv')
movies
```

```python
# Merge the users and movies Dataset
df = pd.merge(users,movies, on='item_id')
df
```

```python
# Exploratory Data Analysis
# 1. Information
df.info()
```

```python
# 2. Maximium and Minimum Rating and Avearage Rating
df['rating'].max()
df['rating'].min()
df['rating'].mean()
df['title'].count()
```

```python
# 3. Statistical Analysis
# Total Users: 943
# Total Movies: 100003
df.describe()
```

```python
# 4.Checking the Empties
df.isnull().sum()
```

```python
# Recommendation based on:
# 1. Rating of the Movies
# 2. Number of Users(Most Viewed Movies)
```

```python
# Grouping all the Movies Titles According to the Average Rating
mean_ratings = df.groupby('title')['rating'].mean()
mean_ratings.sort_values(ascending=False)
```

```python
# Highest Rated Movies
mean_ratings = df.groupby('title')['rating'].mean()
mean_ratings.sort_values(ascending=False).head(10)
```

```python
# Least 5 Rated Movies
mean_ratings = df.groupby('title')['rating'].mean()
mean_ratings.sort_values(ascending=True).head(5)
```

```python
# How many Users Rated a Particular Movie Item(Popularity)
rating_count = df.groupby('title')['rating'].count()
rating_count.sort_values(ascending=False).head(10)
```

```python
# Least 5 popular movies
rating_count = df.groupby('title')['rating'].count()
rating_count.sort_values(ascending=True).head(5)
```

```python
# Create a DataFrame that will Contain, the title, mean_rating, number_of_views
df_new = pd.DataFrame(mean_ratings)
df_new
df_new['number_of_views'] = pd.DataFrame(rating_count)
df_new.sort_values(by='number_of_views', ascending=False)
```

```python
# PIVOT TABLE
# Data can be stored in 3D
# X-Columns
# Y-indexing
# Z-values
```

```python
pd.set_option('display.max_columns', 100)
pivot = df.pivot_table(index='user_id', columns='title', values='rating')
pivot
```

```python
selected_movie = pivot['Godfather, The (1972)']
selected_movie
```

```python
# Recommend a Similar Movie
#Correlation: The similar movie needs to have a good positive correlation with the selected movie
# i.e, If the ratin and users increase in the selected movie, it should also increase in the similar movie
import warnings
warnings.filterwarnings('ignore')

similar = pivot.corrwith(selected_movie)
similar.sort_values(ascending=False).head(10)
```

```python
similar_movies = pd.DataFrame(similar, columns=['correlation'])
similar_movies.sort_values(by='correlation', ascending=False)
similar_movies.sort_values(by='correlation', ascending=False).head(10)
```

```python
# Join the similar_movies with the number of views
similar_movies = similar_movies.join(df_new['number_of_views'])
similar_movies.head(10)
```

```python
condition = similar_movies['number_of_views'] > 200
similar_movies = similar_movies[condition]
recommended_movies = similar_movies.sort_values(by='correlation', ascending=False)
recommended_movies.head(20)
```
